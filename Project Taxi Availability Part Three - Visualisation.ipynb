# Visualization


We visualize the data so that we can detect any possible outliers, anomalies, and build intuition around the data before we perform machine learning modelling in Part IV.  

In this notebook, 
1. Visualize the taxi availability over the month
2. Visualize the taxi availability over the 24 hours in a day
3. Visualize the taxi availability by day of the week




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('taxi_2.csv',index_col=0,parse_dates=[0])
print(df)
df.head()
```

```python
#  Plot properties.taxi_count vs time
plt.figure(figsize=(16, 6))

sns.lineplot(x=df.index,y = "properties.taxi_count",data=df)

```

```python
# Plot suggests that there are 0 taxis available in Singpapore on certain days. This is not possible.
# Get the slice where there is no taxi counts
df[df['properties.taxi_count']==0]
```

```python
# Looks like there was some downtime on those days, as such we will remove them
# Remove rows where properties.taxi_count is 0
df = df[df['properties.taxi_count']!=0]
print(len(df))
```

```python
# Resample DataFrame by day, and take the day's median value, to reduce intra-day variation
rdf = df.resample('D').median()
rdf.head()
```

```python
# Plot resampled DataFrame over original plot
plt.figure(figsize=(16, 6))
sns.lineplot(x=df.index,y = "properties.taxi_count",data=df)
sns.lineplot(x=rdf.index,y="properties.taxi_count",data=rdf)
```
